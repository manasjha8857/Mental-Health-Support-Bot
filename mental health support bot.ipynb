```python
from gtts import gTTS
from textblob import TextBlob
import json
import datetime
from IPython.display import Audio

# Function to generate and play voice-based response
def speak_and_play(text):
    tts = gTTS(text)
    tts.save('response.mp3')
    return Audio('response.mp3', autoplay=True)

# Function to analyze sentiment
def analyze_emotion(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    return sentiment

# Function to identify the type of problem and provide solutions
def problem_solver(user_input):
    problems = {
        "stress": "It seems like you're stressed. Try some relaxation techniques like deep breathing or meditation.",
        "relationship": "Relationships can be difficult. Open communication and understanding are key. Try expressing your thoughts calmly.",
        "work": "Work can be overwhelming. Break your tasks into small steps, and take short breaks to recharge.",
        "motivation": "It seems like you're lacking motivation. Set smaller, achievable goals, and celebrate your successes.",
        "health": "Health is important. Ensure you're getting enough sleep, eating well, and exercising regularly.",
        "anxiety": "Anxiety can be tough. Grounding techniques such as deep breathing or focusing on your surroundings can help.",
        "loneliness": "Feeling lonely? Try reaching out to friends, family, or joining a social group or hobby class.",
        "sleep": "Sleep is vital. Create a relaxing bedtime routine, reduce screen time before bed, and try calming activities like reading.",
        "financial": "Financial worries are common. Make a budget, track your expenses, and seek financial advice if needed.",
        "family": "Family issues can be challenging. Listen to others, express your thoughts calmly, and try to empathize with their feelings.",
        "decision-making": "Decisions can be hard. Try weighing the pros and cons of each option and consider long-term effects.",
        "confidence": "If you're feeling a lack of confidence, focus on your strengths and past achievements. Believe in yourself!",
        "friendship": "Friendship troubles can hurt. Communicate openly with your friends and work to resolve misunderstandings.",
        "grief": "Grief is a difficult process. Take time to feel your emotions and talk to loved ones or a counselor for support.",
        "change": "Change can be unsettling. Try to focus on the positive aspects and opportunities for growth.",
        "anger": "Feeling angry? Take a moment to breathe deeply, step away from the situation, and return when you're calmer.",
        "fear": "Fear can paralyze us. Face it gradually, step by step, and remember that you're stronger than you think.",
        "failure": "Failure is a part of growth. Learn from your mistakes, and remember that every setback is an opportunity to improve.",
        "guilt": "Guilt can weigh heavily. Acknowledge your feelings, make amends if needed, and work towards forgiving yourself."
    }

    # Identify the problem based on the user's input
    for keyword, solution in problems.items():
        if keyword in user_input.lower():
            return solution
    
    return "I'm here for you. If you ever need to talk, just let me know, and we can work through it together."

# Function to store conversation
def store_conversation(data, file_path='conversation_data.json'):
    try:
        # Load existing data
        with open(file_path, 'r') as file:
            conversation_data = json.load(file)
    except FileNotFoundError:
        # If the file doesn't exist, create a new dictionary
        conversation_data = {}

    # Add new conversation with timestamp
    timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    conversation_data[timestamp] = data

    # Save the updated data
    with open(file_path, 'w') as file:
        json.dump(conversation_data, file, indent=4)

# Function to generate a response based on emotion and problem type
def generate_response(emotion_score, user_input, name):
    # Analyze the context for a specific solution
    solution = problem_solver(user_input)

    # Generate a response based on emotion score
    if emotion_score > 0.5:
        response = f"I'm so glad you're feeling good, {name}! Keep up the positive energy! {solution}"
    elif 0 < emotion_score <= 0.5:
        response = f"It seems like you're doing okay, {name}, but maybe something's on your mind. {solution}"
    elif -0.5 <= emotion_score <= 0:
        response = f"I sense you're feeling a bit down, {name}. That's perfectly okay. {solution}"
    else:
        response = f"I'm really sorry you're feeling this way, {name}. {solution} Remember, you're not alone, and things will get better."

    return response

# Main function to interact with the user
def talk_to_me():
    # Ask for the user's name and greet them
    name = input("What's your name? ")
    print(f"Hi {name}, I'm here to listen to anything you want to talk about. How are you feeling today?")
    display(speak_and_play(f"Hi {name}, I'm here to listen to anything you want to talk about. How are you feeling today?"))
    
    while True:
        user_input = input("\nYou: ")
        
        # End the program if the user says "thank you"
        if "thank you" in user_input.lower():
            farewell_message = f"Thank you for using this service, {name}. How can I help you next time?"
            print(farewell_message)
            display(speak_and_play(farewell_message))
            break

        # Analyze sentiment
        emotion_score = analyze_emotion(user_input)

        # Store the conversation
        conversation_entry = {'user_input': user_input, 'emotion_score': emotion_score}
        store_conversation(conversation_entry)

        # Generate a response based on input and emotion
        response = generate_response(emotion_score, user_input, name)

        # Print the response and speak it out loud
        print(f"\nResponse: {response}")
        display(speak_and_play(response))

# Run the conversation
if __name__ == "__main__":
    talk_to_me()
```
